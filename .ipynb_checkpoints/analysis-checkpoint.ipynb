```python
hypertension_count = data.loc[~data.comorbidities.isnull()].comorbidities.apply(lambda x : x.find("hypertension") > -1).sum()
diabetes_count = data.loc[~data.comorbidities.isnull()].comorbidities.apply(lambda x : x.find("diabetes") > -1).sum()
print("Hypertension : {} ({}%)".format(hypertension_count, round(hypertension_count / (data.shape[0] - data.comorbidities.isnull().sum()) * 100,2)))
print("Diabetes : {} ({}%)".format(diabetes_count, round(diabetes_count / (data.shape[0] - data.comorbidities.isnull().sum()) * 100,2)))
```

```python
baseline = ["tac", "mycophenolate", "rapamycin", "pred"]

immunosuppressed = []
non_immunosuppressed = []

for i in range(data.shape[0]):
    row = data.iloc[i,:]
    nochange = True
    allstop = True
    for b in baseline:
        if row[b] == 1:
            if row[b + "_mod"] != "continue":
                nochange = False
            if str(row[b + "_mod"]).find("suspend") == -1:
                allstop = False
    if nochange:
        immunosuppressed += [i]
    if allstop:
        non_immunosuppressed += [i]
non_immunosuppressed = data.iloc[non_immunosuppressed, :]
immunosuppressed = data.iloc[immunosuppressed, :]

            
```

```python
print(immunosuppressed.gender.value_counts())
print("Mean Age: {}".format(immunosuppressed.age.mean()))
print("Minimum Age: {}".format(immunosuppressed.age.min()))
print("Maximum Age: {}".format(immunosuppressed.age.max()))

print(non_immunosuppressed.gender.value_counts())
print("Mean Age: {}".format(non_immunosuppressed.age.mean()))
print("Minimum Age: {}".format(non_immunosuppressed.age.min()))
print("Maximum Age: {}".format(non_immunosuppressed.age.max()))

plt.bar([0,1], immunosuppressed.gender.value_counts() / immunosuppressed.shape[0] * 100, width=0.4, label="immunosuppressed")
plt.bar([0.4,1.4], non_immunosuppressed.gender.value_counts() / non_immunosuppressed.shape[0] * 100, width=0.4, label="non-immunosuppressed")
plt.ylabel("Percentage")
plt.xlabel("Gender")
plt.xticks([0.2,1.2],["M","F"])
plt.legend()
```
